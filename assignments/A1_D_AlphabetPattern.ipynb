```python
#  Pattern 1(D)

print("\n--- Pattern ---") 
rows_d = 4  
cols_d = 3   
 
for row_number in range(1, rows_d + 1): 
    for column_number in range(1, cols_d + 1): 
        if (row_number == 1 or row_number == rows_d or 
            column_number == 1 or column_number == cols_d): 
            print("* ", end="") 
        else: 
            print("  ", end="") 
    print()
```
