```python
# word frequency count Program

text = input("Text: ")
print("Text is", text)
```

```python
clean_text = ""
for char in text:
    if 'a' <= char <= 'z' or 'A' <= char <= 'Z' or char == ' ':
        clean_text += char  
clean_text = clean_text.lower()
words = clean_text.split()
word_counts = {} 
for word in words:
    if word: 
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1
print(word_counts)
```
