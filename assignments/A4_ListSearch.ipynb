```python
item = int(input("Enter item to search: "))
print(f"Searching for {item} in the list...")
```

```python
# Program to search for an item in a list

numbers = [12, 45, 33, 21, 88]

found = False 

for i in range(len(numbers)):
    if numbers[i] == item:
        print(f"Item found at position {i + 1}")
        found = True
        break

if not found:
    print("Item not found.")
```
