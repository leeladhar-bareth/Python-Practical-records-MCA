```python
# Dice and Coin Selection Program

print("Choose an option:")
print("1. Roll a Dice")
print("2. Flip a Coin")
choice = int(input("Enter your choice (1 or 2): "))
print("You selected option:", choice)
if choice == 1:
    face = int(input("Enter a dice face (1 to 6): "))
    print("You entered dice face:", face)
elif choice == 2:
    outcome = input("Enter outcome (H for Head, T for Tail): ").upper()
    print("You entered outcome:", outcome)
else:
    print("Invalid choice!")
```

```python
if choice == 1:
    if 1 <= face <= 6:
        print("Probability of getting", face, "=", 1/6)
    else:
        print("Invalid face! Please enter a value between 1 and 6.")
elif choice == 2:
    if outcome == "H" or outcome == "T":
        print("Probability of getting", outcome, "=", 1/2)
    else:
        print("Invalid input! Please enter H or T.")
else:
    print("Invalid choice! Select 1 or 2.")
```
