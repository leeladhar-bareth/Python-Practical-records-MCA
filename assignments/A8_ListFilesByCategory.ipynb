```python
# Python program to list files and categorize them by extensions

files = ['report.pdf', 'data.csv', 'photo.jpg', 'summary.pdf', 'script.py']

categorized = {}   

for file in files:
   
    name, ext = file.rsplit('.', 1)
    
    if ext not in categorized:
        
        categorized[ext] = []

    categorized[ext].append(file)

print(categorized)
```
